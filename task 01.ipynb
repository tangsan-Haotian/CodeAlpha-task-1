```python
import subprocess

def get_connected_devices():
    try:
        # Run arp-scan command
        result = subprocess.run(["sudo", "arp-scan", "-l"], capture_output=True, text=True)

        # Check if arp-scan ran successfully
        if result.returncode != 0:
            print("Error executing arp-scan. Make sure it's installed and try running with sudo.")
            return []

        # Process the output to find IP and MAC addresses
        devices = []
        lines = result.stdout.splitlines()
        for line in lines[2:-3]:  # Skip headers and footers
            parts = line.split()
            if len(parts) >= 2:
                ip = parts[0]
                mac = parts[1]
                devices.append({"IP": ip, "MAC": mac})

        return devices
    except Exception as e:
        print(f"An error occurred: {e}")
        return []

# Display the connected devices
connected_devices = get_connected_devices()
print("Connected Devices:")
for device in connected_devices:
    print(f"IP: {device['IP']}, MAC: {device['MAC']}")
```
